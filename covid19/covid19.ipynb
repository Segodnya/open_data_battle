```python
import pandas as pd
import numpy as np

def filter_rows_by_values(df, col, values):
    return df[~df[col].isin(values)]


covid = pd.read_csv('covid-19cases.csv')

# Произвести очистку данных, убрав пустые столбцы
# и неинформативные строки (строки, выделенные для заметок for Notes)
covid.drop(['Unnamed: 0', 'Unnamed: 40', 'Unnamed: 41'], axis=1, inplace=True)
covid = covid[covid['date'] != 'For Notes: ']

covid.head()
```

```python
covid.info()
```

```python
# Определить статистику случаев заражения в разных частях света
# (столбец continent) за каждый месяц

table = pd.pivot_table(covid, values='new_cases', index=['month'],
                       columns=['continent'], aggfunc=np.sum)

table
```

```python
# Определить корреляцию каждой части света с другой.

corr = table.corr()
corr
```

```python
# Вывести название части света, которая меньше всего коррелирует с другими

corr.describe()
```
